### First we'll load in the 'Energy and Mining' data obtained from the World Bank data 
__Many of the "countries" in the data set are actually regions, or groups of countries, and are unneccessary for this analysis, so those were removed by country code.__

```python
data=pd.read_csv('energy-and-mining.csv')
not_countries=['WLD','HIC','OED','PST','IBT','LMY','MIC','NAC','IBD',
              'UMC','EUU','EAS','LTE','EMU','ECS','EAR','EAP','TEA',
              'HPC','MNA','TSS','SSF','TMN','SSA','TSA','IDB','IDX',
              'CEB','TEC','PRE','LAC','FCS','LMC','TLA','LDC','MEA',
              'LIC','IDA','ECA','PSS','OSS','CSS','LCN','ARB','SST'
              ,'SAS']
data=data[~data['Country Code'].isin(not_countries)]
data=data.rename(columns={'Country Name':'Country'})
data
```

__Unfortunately, the previous dataset was missing **total** CO2 emission, so we need a separate dataset for that, also obtained from World Bank. This comes in the form of a pivot table, so we'll melt this into a dataframe that will be easier to use in Vega Lite.__

```python
os.chdir('/Users/nickbohmann/Documents/uw/dataviz/A2_stuff/World-Bank-Data-by-Indicators/engergy-and-mining')
data_new=pd.read_csv('data_new.csv')
data_new=data_new[~data_new['Country Code'].isin(not_countries)]
data_new=data_new.drop(columns=['Indicator Code','Indicator Name','Country Code'])
data_new=pd.melt(data_new, 
            id_vars='Country Name', 
            value_vars=list(data_new.columns[1:]), # list of days of the week
            var_name='Year', 
            value_name='CO2 Emission')
data_new=data_new[~data_new.Year.str.contains('Unnamed: 65')]
data_new.to_csv('tot_co2.csv')
data_new
```
